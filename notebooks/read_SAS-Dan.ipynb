```python
import pandas as pd
import os
import re
```

```python
years={"_E":"_0708",
       "_J":"_1718"}
```

```python
files_df=os.listdir("../sas")
files_df.remove("columnreader")
names_df=[]
for file in files_df:
    for key, value in years.items():
        file=file.replace(".XPT","").replace(key,value)
    names_df.append(file)
print(files_df)
print(names_df)
```

```python
df={}
for name,file in zip(names_df,files_df):
    df[name]=pd.read_sas(f"../sas/{file}")
df["CBQ_0708"].head()
```

```python
files_col=os.listdir("../sas/columnreader")
texts_col={}
names_col=[]
for file in files_col:
    for key, value in years.items():
        file=file.replace(".txt","").replace(key,value)
    names_col.append(file)

if names_col!=names_df:
    print("SOMETHING'S WRONG")
else:
    print("Columns match the dataframes:")
    print(files_col)
    print(names_col)
```

```python
texts={}
for name,file in zip(names_col,files_col):
    with open(f"../sas/columnreader/{file}", "r") as f:
        texts[name]=f.read().split("\n")
texts[names_col[3]][0:5]
```

```python
column_renamer={}
for name in names_col:
    text_list=texts[name]
    line_dict={}
    print(text_list)
    for line in text_list:
        split_line=re.findall("([A-z0-9]*) - ([#'A-z \/0-9\-?]*)",line)
        line_dict[split_line[0][0]]=split_line[0][1].replace(" - ","_").replace(" ","_").replace("-","_").replace("'s","").replace("/", "_").replace("?","").replace("#","Num")
    column_renamer[name]=line_dict
    
column_renamer[names_col[0]]
```

```python
for name in names_df:
    df[name]=df[name].rename(columns=column_renamer[name])
list(df[names_df[0]].columns)
```

```python
merged_df={}
for key,value in years.items():
    year_tables=list(filter(lambda x:value in x,names_df))
    df_merge=df[year_tables[0]]
    for i,table in enumerate(year_tables):
        if i > 0:
            df_merge=df_merge.merge(df[table],on="Respondent_sequence_number",how="inner")
    merged_df[value]=df_merge
merged_df["_1718"]
```
