```python
import re
line_dict={}
for line in text_list:
    split_line=re.findall("([A-z0-9]*) - ([#'A-z \/0-9\-?]*)",line)
    line_dict[split_line[0][0]]=split_line[0][1].replace(" - ","_").replace(" ","_").replace("-","_").replace("'s","")
line_dict

```
